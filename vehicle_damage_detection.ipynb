```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
model = YOLO('models/best.pt')
model
```

```python
img_path = 'test.jpg'
results = model(img_path, save=True)
output_img_path = results[0].save_dir + '/' + img_path.split('/')[-1]
img = Image.open(output_img_path)
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
```

```python
results = model(img_path)
for box in results[0].boxes:
    print('Class:', int(box.cls[0]))
    print('Confidence:', float(box.conf[0]))
    print('Bounding Box:', box.xyxy[0].tolist())
    print('---')
```

```python
import glob
image_list = glob.glob('test_images/*')
for path in image_list:
    print('Processing:', path)
    results = model(path, save=True)
```

```python
model = YOLO('yolov8n.pt')
model.train(
    data='dataset/data.yaml',
    epochs=50,
    imgsz=640,
    batch=8
)
```

```python
metrics = model.val(data='dataset/data.yaml')
print(metrics)
```

```python
import cv2
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    results = model(frame)
    annotated_frame = results[0].plot()
    cv2.imshow('Damage Detection', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
